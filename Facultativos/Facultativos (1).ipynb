Este m√©todo, conocido como "calificaci√≥n por exposici√≥n", es crucial en situaciones donde el reasegurador no dispone de un historial de siniestros cre√≠ble suficiente de la parte asegurada.

Limitaciones y cr√≠ticas:

A pesar de su amplia relevancia, las tablas de Salzmann presentaban limitaciones importantes:

**Antig√ºedad de los Datos:**

  Las curvas originales se basaron en datos de siniestros por incendio de viviendas de 1960. Esto significa que los valores num√©ricos espec√≠ficos de las tablas estaban desactualizados debido a los cambios en las pr√°cticas de construcci√≥n, c√≥digos de edificaci√≥n, tecnolog√≠as de protecci√≥n contra incendios, valores de las propiedades y el panorama general de riesgos a lo largo del tiempo.

**Intenci√≥n Original de Salzmann:**

 La propia Salzmann concibi√≥ sus curvas "solo como un ejemplo" para demostrar su metodolog√≠a y expl√≠citamente "no las recomend√≥ para uso posterior" como herramientas de tarificaci√≥n universales. Esto indica que su uso prolongado sin actualizaciones o ajustes podr√≠a haber llevado a inexactitudes.

**Diferencia entre Datos y Aplicaci√≥n:**

 Los datos de siniestros utilizados en su an√°lisis original "difieren significativamente de lo que t√≠picamente cubre un contrato de reaseguro de exceso de p√©rdida de propiedad".


**Dependencia de Datos Externos:**

  El m√©todo de calificaci√≥n por exposici√≥n requiere que el reasegurador asigne un alto grado de credibilidad a fuentes de datos que no son propias, dependiendo de datos derivados de otras aseguradoras y sistemas de calificaci√≥n de terceros.

**Aproximaciones:**
   Una desventaja del m√©todo es que pueden haber porcentajes de retenci√≥n donde las p√©rdidas se aproximan.




Salzmann tables were primarily used for rating non-proportional property excess of loss reinsurance contracts. This method, known as "exposure rating", is crucial in situations where the reinsurer does not have a sufficient credible loss history from the insured party.

Limitations and critics:
Despite their wide relevance, the Salzmann tables had significant limitations:

**Data Age:**

 The original curves were based on residential fire loss data from 1960. This means that the specific numerical values in the tables were outdated due to changes in construction practices, building codes, fire protection technologies, property values, and the overall risk landscape over time.

**Salzmann's Original Intent:**

Salzmann herself conceived her curves "only as an example" to demonstrate her methodology and explicitly "did not recommend them for further use" as universal rating tools. This indicates that their prolonged use without updates or adjustments could have led to inaccuracies.

**Difference Between Data and Application:**

The loss data used in her original analysis "differs significantly from what is typically covered by a property excess of loss reinsurance treaty."

**Dependence on External Data:**

 The exposure rating method requires the reinsurer to assign a high degree of credibility to data sources that are not their own, relying on data derived from other insurers and third-party rating systems.

**Loss Zone:**

A disadvantage of the method is that it creates a zone in each layer where losses approach, but do not reach, the next level of retention.

```python
# Listas v√°lidas para validaci√≥n
CLASES_VALIDAS = ["Proporcional", "No proporcional"]
TIPOS_VALIDOS = [
   "QS", "SPL", "XL por riesgo", "XL por acontecimiento",
   "XL por Cat√°strofe", "Stop Loss", "Stop Loss por ejercicio"
]

# Crear DataFrame vac√≠o si no existe
try:
   contratos_reaseguro
except NameError:
   contratos_reaseguro = pd.DataFrame(columns=[
    "id_contrato", "nombre_contrato", "clase_contrato", "tipo_contrato",
    "fecha_inicio", "fecha_fin", "retencion", "limite", "capacidad",
    "afecta_facultativo", "activo", "reaseguradora", "comentarios"
   ])

def cargar_contrato(
   df=contratos_reaseguro,
   id_contrato=None,
   nombre_contrato=None,
   clase_contrato=None,
   tipo_contrato=None,
   fecha_inicio=None,
   fecha_fin=None,
   retencion=0.0,
   limite=0.0,
   afecta_facultativo=False,
   reaseguradora=None,
   comentarios=""
):
   """
   Carga un contrato de reaseguro en el DataFrame con validaciones.

   Par√°metros:
    df: DataFrame destino (por defecto contratos_reaseguro)
    id_contrato: ID √∫nico del contrato
    nombre_contrato: Nombre o alias del contrato
    clase_contrato: "Proporcional" o "No proporcional"
    tipo_contrato: Tipo espec√≠fico (ej. "QS", "XL por riesgo", etc.)
    fecha_inicio: Fecha de inicio en formato "dd/mm/yyyy"
    fecha_fin: Fecha de fin en formato "dd/mm/yyyy"
    retencion: Monto de retenci√≥n (no negativo)
    limite: Monto de l√≠mite (no negativo)
    afecta_facultativo: Booleano si afecta al facultativo
    reaseguradora: Nombre de la reaseguradora
    comentarios: Texto adicional

   Retorna:
    DataFrame actualizado con el nuevo contrato
   """

   # Validar clase de contrato
   if clase_contrato not in CLASES_VALIDAS:
    raise ValueError(f"Clase de contrato inv√°lida. Debe ser una de: {CLASES_VALIDAS}")

   # Validar tipo de contrato
   if tipo_contrato not in TIPOS_VALIDOS:
    raise ValueError(f"Tipo de contrato inv√°lido. Debe ser uno de: {TIPOS_VALIDOS}")

   # Validar fechas y convertir a datetime
   try:
    fecha_inicio_dt = datetime.strptime(fecha_inicio, "%d/%m/%Y")
    fecha_fin_dt = datetime.strptime(fecha_fin, "%d/%m/%Y")
   except ValueError:
    raise ValueError("Formato de fecha inv√°lido. Use 'dd/mm/yyyy'.")

   if fecha_inicio_dt >= fecha_fin_dt:
    raise ValueError("La fecha de inicio debe ser anterior a la fecha de fin.")

   # Validar retenci√≥n y l√≠mite
   if retencion < 0 or limite < 0:
    raise ValueError("Retenci√≥n y l√≠mite deben ser valores no negativos.")

   # Calcular capacidad
   capacidad = retencion + limite

   # Determinar si el contrato est√° activo (vigente hoy)
   hoy = datetime.today()
   activo = fecha_inicio_dt <= hoy <= fecha_fin_dt

   # Crear nuevo registro
   nuevo_contrato = {
    "id_contrato": id_contrato,
    "nombre_contrato": nombre_contrato,
    "clase_contrato": clase_contrato,
    "tipo_contrato": tipo_contrato,
    "fecha_inicio": fecha_inicio_dt,
    "fecha_fin": fecha_fin_dt,
    "retencion": retencion,
    "limite": limite,
    "capacidad": capacidad,
    "afecta_facultativo": afecta_facultativo,
    "activo": activo,
    "reaseguradora": reaseguradora,
    "comentarios": comentarios
   }

   # Agregar al DataFrame
   df = pd.concat([df, pd.DataFrame([nuevo_contrato])], ignore_index=True)

   return df
```

#Graficos

Pasos a seguir: 1ro calcular la cantidad de capas para el grafico, 2do calcular la participacion dentro de cada capa.

Paso3:

Liquidar siniestros: Ver cu√°l es el monto de siniestros y distribuirlos seg√∫n el gr√°fico. por capas. Para aquellas capas que tienen m√°s de una aseguradora, distribuir el monto correspondientee a esa capa  seg√∫n el porcentaje de participaci√≥n de esa capa.


```python
import pandas as pd
import matplotlib.pyplot as plt
from openpyxl import load_workbook
from openpyxl.drawing.image import Image as XLImage

# Datos de participaci√≥n por capa y aseguradora. Editarlo para que sea la cantidad respectiva del programa.
data = {
    'Capa': ['1st Layer', '2nd Layer', '3rd Layer', '4th Layer'],
    'Desde': [0, 1_000_000, 3_000_000, 6_000_000],
    'Hasta': [1_000_000, 3_000_000, 6_000_000, 10_000_000],
    'Aseguradora A': [0, 2_000_000, 3_000_000, 2_000_000],
    'Aseguradora B': [0, 0, 0, 2_000_000],
    'Deducible': [1_000_000, 0, 0, 0]
}

df = pd.DataFrame(data)

List_participantes_programa = [col for col in df.columns if col not in ['Capa', 'Desde', 'Hasta']]




# Altura total de la suma asegurada
total_sa = 10_000_000
```

```python

# Crear gr√°fico
fig, ax = plt.subplots(figsize=(6, 8))

# Para evitar etiquetas duplicadas en la leyenda
legend_labels = {}

# Colores por aseguradora
colors = {
    'Aseguradora A': '#1f77b4',
    'Aseguradora B': '#ff7f0e',
    'Deducible': '#2ca02c'
}

# Dibujar cada capa como una caja de igual altura
for i, row in df.iterrows():
    capa_altura = row['Hasta'] - row['Desde']
    bottom = row['Desde']
    total_participacion = sum(row[aseg] for aseg in List_participantes_programa)
    if total_participacion == 0:
        continue
    left = 0

    for aseguradora in List_participantes_programa:
        participacion = row[aseguradora]
        if participacion > 0:
            ancho = participacion / total_participacion
            bar = ax.barh(
                y=(bottom + capa_altura / 2),
                width=ancho,
                height=capa_altura,
                left=left,
                color=colors[aseguradora],
                edgecolor='black'
            )
            # A√±adir etiqueta con el nombre de la aseguradora
            ax.text(
                x=left + ancho / 2,
                y=(bottom + capa_altura / 2),
                s=aseguradora,
                va='center',
                ha='center',
                fontsize=8,
                color='white',
                weight='bold'
            )
            if aseguradora not in legend_labels:
                legend_labels[aseguradora] = bar[0]
            left += ancho


# Configurar eje y
ax.set_ylim(0, total_sa)
ax.set_xlim(0, 1)
ax.set_xlabel('Participaci√≥n proporcional')
ax.set_ylabel('Suma Asegurada')
ax.set_title('Distribuci√≥n de Capas y Participaci√≥n por Aseguradora')
ax.set_yticks([0, 1_000_000, 3_000_000, 6_000_000, 10_000_000])
ax.set_yticklabels(['0', '1M', '3M', '6M', '10M'])

# Agregar leyenda con etiquetas √∫nicas
ax.legend(legend_labels.values(), legend_labels.keys(), loc='upper right')

# Guardar gr√°fico como imagen
plt.tight_layout()
plt.savefig("grafico_facultativo.png")


# Guardar datos y explicaci√≥n en Excel
with pd.ExcelWriter("esquema_facultativo.xlsx", engine='openpyxl') as writer:
    df.to_excel(writer, sheet_name='Datos', index=False)

    instrucciones = [
        "INSTRUCCIONES PARA EDITAR Y ENTENDER EL GR√ÅFICO",
        "",
        "1. Cada fila en la hoja 'Datos' representa una capa del contrato.",
        "2. Las columnas 'Desde' y 'Hasta' indican el intervalo de suma asegurada cubierto por esa capa.",
        "3. Las columnas de aseguradoras indican cu√°nto cubre cada una dentro de esa capa.",
        "4. El gr√°fico muestra una barra vertical de 0 a 10 millones.",
        "5. Cada capa tiene la misma altura visual, y si hay m√°s de una aseguradora, se divide horizontalmente.",
        "6. Para modificar el gr√°fico, edite los valores en la hoja 'Datos' y regenere el gr√°fico con el script."
    ]
    instrucciones_df = pd.DataFrame({'Instrucciones': instrucciones})
    instrucciones_df.to_excel(writer, sheet_name='Instrucciones', index=False)

# Cargar el archivo Excel y agregar el gr√°fico como imagen en una nueva hoja
wb = load_workbook("esquema_facultativo.xlsx")
ws = wb.create_sheet("Gr√°fico")
img = XLImage("grafico_facultativo.png")
ws.add_image(img, "A1")
wb.save("esquema_facultativo.xlsx")
```

#Liquidaci√≥n de Siniestros

```python
import pandas as pd

# Datos de participaci√≥n por capa y aseguradora. Editarlo para que sea la cantidad respectiva del programa.
data = {
    'Capa': ['1st Layer', '2nd Layer', '3rd Layer', '4th Layer'],
    'Desde': [0, 1_000_000, 3_000_000, 6_000_000],
    'Hasta': [1_000_000, 3_000_000, 6_000_000, 10_000_000],
    'Aseguradora A': [0, 2_000_000, 3_000_000, 2_000_000],
    'Aseguradora B': [0, 0, 0, 2_000_000],
    'Deducible': [1_000_000, 0, 0, 0]
}

df = pd.DataFrame(data)

# Detectar columnas num√©ricas y aplicar formato
formato = {col: '{:,.0f}' for col in df.select_dtypes(include='number').columns}
df.style.format(formato)



List_participantes_programa = [col for col in df.columns if col not in ['Capa', 'Desde', 'Hasta']]

# Siniestro total
siniestro_total = 9_500_000

# Inicializar resultados
resultados = []

# Recorrer cada capa
for i, row in df.iterrows():
    capa_desde = row['Desde']
    capa_hasta = row['Hasta']
    monto_capa = min(siniestro_total, capa_hasta) - capa_desde
    if monto_capa <= 0:
        continue

    total_participacion = sum(row[aseg] for aseg in List_participantes_programa)

    if total_participacion == 0:
        continue

    capa_resultado = {'Capa': row['Capa'], 'Monto Capa': monto_capa}
    for aseguradora in List_participantes_programa:
        participacion = row[aseguradora]
        monto_pagado = monto_capa * (participacion / total_participacion)
        capa_resultado[aseguradora] = monto_pagado
    resultados.append(capa_resultado)

# Convertir resultados a DataFrame
df_resultado = pd.DataFrame(resultados)

# Mostrar resultados
for participante in List_participantes_programa:
    print(f"{participante} se har√° cargo de ${df_resultado[participante].sum():,.0f}")

df_resultado
```
